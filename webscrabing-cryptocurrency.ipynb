```python
from bs4 import BeautifulSoup
import requests
```

```python
website = 'https://www.coingecko.com/en'
response = requests.get(website)
response.status_code
soup = BeautifulSoup(response.content, 'html.parser')
results = soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
```

```python
coin_rank=[]
coin_name = []
coin_symbol=[]
price = []
change_1h = []
change_24h = []
change_7d = []
volume_24h = []
market_cap=[]
```

```python
len(results)
```

```python
for element in range(len(results)):
        coin_rank.append(results[element].find('td', {'class':'table-number tw-text-left text-xs'}).get_text().strip())
        coin_name.append(results[element].find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
        coin_symbol.append(results[element].find('span', {'class':'tw-hidden d-lg-inline font-normal text-3xs ml-2'}).get_text().strip())
        price.append(results[element].find('td', {'class':'td-price price text-right'}).get_text().strip())
        change_1h.append(results[element].find('td', {'class':'td-change1h change1h stat-percent text-right col-market'}).get_text().strip())
        change_24h.append(results[element].find('td', {'class':'td-change24h change24h stat-percent text-right col-market'}).get_text().strip())
        change_7d.append(results[element].find('td', {'class':'td-change7d change7d stat-percent text-right col-market'}).get_text().strip())
        volume_24h.append(results[element].find('td', {'class':'td-liquidity_score'}).get_text().strip())
        market_cap.append(results[element].find('td', {'class':'td-market_cap'}).get_text().strip())
```
